# Get Stock all available stock prices via web.DataReader (Almost ALL)

```python
import pandas_datareader.data as web
import datetime as dt
start = dt.datetime(2000,1,1)
end = dt.datetime(2020,12,31)
end_ = dt.datetime.now()

panel_data = web.DataReader('SPY', 'yahoo', start, end_)

print("panel_data:",panel_data)

import bs4 as bs
import requests
resp = requests.get('https://stockanalysis.com/stocks/')
soup = bs.BeautifulSoup(resp.text)

```

```python
em = []
for i in soup.find('ul', class_ = 'no-spacing'):
    a = i.text.split('-')[0]
    a = a.replace(' ','')
    em.append(a)
```

```python
print(em)
```

```python
# df = web.DataReader('AAC', 'yahoo', start, end)
em_unable = []
# print(df)

print(em[:3])
for ticker in em:#shorten
    try: 
        print('ticker:', ticker)
        df = web.DataReader(ticker, 'yahoo', start, end)
        df.to_csv('C:/Users/malco/Desktop/Server Data 2/all stock price 2000_to_31Jan2020/{}.csv'.format(ticker))
    except:
        em_unable.append(ticker)
        print("unable:", ticker)
        pass

```

```python
#Write stocks that are unable to retrieve
with open('C:/Users/malco/Desktop/Server Data 2/all stock price 2000_to_31Jan2020_INFO(unretrieved).txt', 'w') as f:
    for i in em_unable:
        f.write(i +'\n')
```

```python
#Write stocks that have been retrieve
retrieved = []
for i in em:
    print(i)
    if i not in em_unable:
        print('not in')
        retrieved.append(i)
print(retrieved)

with open('C:/Users/malco/Desktop/Server Data 2/all stock price 2000_to_31Jan2020_INFO(retrieved).txt', 'w') as f:
    for i in retrieved:
        f.write(i +'\n')
```
